```python
import pandas as pd
```

```python
df2
```

```python
#df2.plot.line()
df2['c'].plot.line(figsize=(10,3),ls=':', c='red',lw=5)
```

```python
title='MY PLOT TITLE'
xlabel ='MY X Data'
ylabel ='My Y Data'
```

```python
ax=df2['c'].plot.line(figsize=(10,3),ls=':', c='red',lw=5)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

```python
ax=df2.plot()
ax.legend(loc=2,bbox_to_anchor=(1.0,1.0))
```
